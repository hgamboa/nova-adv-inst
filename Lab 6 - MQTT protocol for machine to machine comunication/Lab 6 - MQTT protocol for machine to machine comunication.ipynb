## Prerequisites:

### Learn about MQTT
You can find a good MQTT tutorial in: http://www.steves-internet-guide.com/mqtt-basics-course/

In this course, we are going to use the mosquitto implementation of the MQTT protocol. More details in: https://mosquitto.org/

In the laboratory network it is available a MQTT broker, in the IP address (no password is required):
* 192.168.1.97 & 192.168.1.98

### Install MQTT services in the RPI
To install the necessary MQTT services on your Raspberry Pi, open the terminal console and run the following commands:
* sudo apt install -y mosquitto mosquitto-clients
* sudo systemctl enable mosquitto.service
  
Test if the services are properly installed and running, by doing:
* sudo systemctl status mosquitto

### Install paho-mqtt library for python
To integrate MQTT functionality into your Python programs, you need to install the appropriate library: paho-mqtt.

#### In the RPI
* From the RPI terminal console, run:
  * sudo pip install paho-mqtt
* Complementary information in: https://pypi.org/project/paho-mqtt/

#### In the Anaconda (PC)
* Open anaconda prompt and confirm that you have the right enviorment active. Run the following command:
  * conda install -c conda-forge paho-mqtt
* Complementary information in: https://anaconda.org/conda-forge/paho-mqtt

## Goal 1 - Publish and subscribe MQTT messages

The MQTT is a publish/subscribe messaging protocol to share data between machines. Two actions are possible, send your data to a broker, by publishing it in a topic or receive it by subscribing a topic. So, to test your MQTT service you need the following information:
* broker IP address – in the lab: 192.168.1.97 or 192.168.1.98
* topic to be published or subscribed – Your choice, but it must be unique. 
* data load – could any data, but if it is binary data, it must be serialized in a string. For more information see json formats.

### Publish data

* Open a termincal console in your RPI and run:
  * mosquitto_pub -h 192.168.1.98 -t "AAIB/g0/mytest" -m "Hello World"
 
This action has publisded the data "Hello World" in the topic "AAIB/g0/mytest". 


### Subscribe data

* Open a terminal console and subscribe the topic, by doing:
   * mosquitto_sub -h 192.168.1.98 -v -t "AAIB/g0/mytest/#"

Note 1: '#' means any subtopif of this topic
Note 2: to terminate the subscription, do Ctrl-C

* Publish more data, has you have done previously and confirm that you are receiving it.

### Change topic

As the broker is unique, if you topic is not unique, you can not distiguish the data  publish by you, from the other users.

* Change the topic, for someting unique to you.
  * subscribe this new topic
  * publish data in this new topic

Final note: If you want, you can see the messages that are being generated by the other groups, if they are kind enough to give you the topic where they are publishing.

### Goal 2.2 - Second scenario

In goal 2.1, the RPI only has the capability to publish data and the PC to subscribe. In a more real scenario, it is often necessary that both machines, PC and RPI, to be able to publish and subscribe simultaneously. That introduce a new layer of complexity, due to the subscribe data could reach the machine at any time, so it necessary to have some ability to do multithreading (running diferent tasks/threads at same time).  

Let’s assume the following scenario:
* The PC publish the character "s" in the topic: "AAIB/anything/cmd" (replace anything with a string that uniquely intentify you)
* The RPI subscribe the topic: "AAIB/anything/cmd"
* When the RPI receives the caracter "s",
  * it generated a list of 10 random numbers
  * publish the list in the topic: "AAIB/anything/data" (replace anything with a string that uniquely intentify you)
* The PC subsctibes the topic: "AAIB/anything/data"
