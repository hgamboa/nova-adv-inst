## Project Based in Image:

### Prerequisites:
In your RPI install the open computer vision library:
* sudo apt update
* sudo apt install python3-opencv
* sudo pip install --upgrade opencv-python

### Sample Software
As a starting point you have two softwares available:
* Lab8_goal_image_PC.py
  * Copy this file to a folder in your computer
  * Since the interface is developed in streamlit, run it in the anaconda prompt, with:
    * streamlit run Lab8_goal_image_PC.py
* Lab8_goal_image_RPI.py
  * Transfer this file to a folder in the RPI
  * Run it with:
    * python Lab8_goal_image_RPI.py

Before running the sofwares, attached the camera to one of the USB ports of the RPI

After both softwares are running:
* Choose a MQTT topic the identify you, and start the MQTT connection
* Request a single frame. You should have now an image in the streamlit interface.
* The acquired imagem is also saved in the folder of the RPI and PC
* The saved image is in color and has a resolution of 640*480, but you can change this parameters.
* To close the programs properly, disconnecting all the hardware properly, press the buttons:
  * Terminate RPI Process
  * MQTT Disconnect

This two programs, togheter with the ones developed in the previous labs, should give you a good starting framework to develop your project.

## Project Based in Inercial Sensors (IMU):

### Prerequisites:
In your RPI install the bluetooth libraries:
* sudo apt update
* sudo pip install bleak

### Sample Software
As a starting point you have three softwares available:
* Lab8_goal_imu_PC.py
  * Copy this file to a folder in your computer
  * Since the interface is developed in streamlit, run it in the anaconda prompt, with:
    * streamlit run Lab8_goal_imu_PC.py
* Lab8_goal_imu_RPI.py
  * Transfer this file to a folder in the RPI
  * Run it with:
    * python Lab8_goal_imu_RPI.py
* Lab8_goal_imu_M5Stick.ino
  * This is the sofware that is pre-installed in the M5-Stick.
  * The M5-Stick has the imu devices (accelarometers and gyroscopes) and bluetooth communication. Since in has also its one battery, you can use it in an autonomous way
  * If you need the modify the provided version of the sofware, please refeer to the older version of the script: Lab8 - M5Stick.pdf 

Before running the sofwares, and since you are using the M5-Stick, power it up first by pressing the left button. A few messages should display in the screen, including the device name. This device name must the same that is in Lab8_goal_imu_RPI.py. Edit this file and confirm.

After both softwares are running:
* Choose a MQTT topic the identify you, and start the MQTT connection
* Start acquisition and move the M5-Stick
* After a couple of seconds, the streamlit will display a plot with the information of the accelarometers 
* The information acquired from the accelarometers and gyroscopes is saved in the folder of the RPI and PC in a csv format
* The imu data was acquired for 2 seconds at a sample rate of 50 samples/s, but you can change this parameters.
* To close the programs properly, disconnecting all the hardware properly, press the buttons:
  * Terminate RPI Process (note: bluetooth sometime is quite hard to disconnect. If there are problems reboot M5-Stick by pressing left button for 3 seconds)
  * MQTT Disconnect

This two programs, togheter with the ones developed in the previous labs, should give you a good starting framework to develop your project.
